```python
from copy import deepcopy

from decorator import append

a = [1,2]
b = a
b.append(3)

print(a, b)
```

```python
def emute(xs: list[int]) -> None:
    xs.append(3)

def rebind(xs: list[int]) -> None:
    xs = [1,2]

li = [1]
emute(li)
print(li)

rebind(li)
print(li)
```

```python
import copy

li = [[1], [2]]
l1 = li.copy()
l2 = copy.deepcopy(li)
l3 = li

l1[0].append(4)
print(li, l1, l2, l3)

```
